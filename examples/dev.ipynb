```python
stations = ['03t', '05t']
measurements=['O3', 'CO', 'NO2', 'PM25']
```

```python
URL = 'http://air4thai.pcd.go.th/webV2/history/api/data.php?stationID={station_ids}' +  \
     '&param={measurements}&type=hr&sdate={sdate}&edate={edate}'+ \
     '&stime=00&etime=24'
```

```python
url = URL.format(station_ids=','.join(stations), measurements=','.join(measurements), sdate='2019-01-01', edate='2019-02-03')
```

```python
print(url)
```

```python
import urllib.request, json 
with urllib.request.urlopen(url) as api:
    data = json.loads(api.read().decode())
    print(data)
```

```python
records = []

for s in data['stations']:
    for r in s['data']:
        r['DATETIMEDATA'] = 
        records.append(dict(stationID=s['stationID'], **r))
```

```python
records[:10]
```

```python
cols = ['stationID', 'DATETIMEDATA'] + measurements
df = pd.DataFrame(records)[cols].sort_values(['stationID', 'DATETIMEDATA'])
df['DATETIMEDATA'] = pd.to_datetime(df['DATETIMEDATA'], format='%Y-%m-%d %H:%M:%S')
```
